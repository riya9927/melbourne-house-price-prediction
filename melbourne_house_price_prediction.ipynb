```python
####loading packages:
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
### ignore warnings:
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.columns
```

```python
###we will work with fewer features to begin with-:
variables_to_use=['Suburb','Rooms','Type','Method','SellerG','Distance','Regionname','Propertycount',
                  'CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price','YearBuilt']

```

```python
dataset=dataset[variables_to_use]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
####missing value check:
dataset.isna().sum()
```

```python
dataset.info()
```

```python
dataset['Bedroom2']=dataset['Bedroom2'].fillna(dataset.Bedroom2.mean())
dataset['Bathroom']=dataset['Bathroom'].fillna(dataset.Bathroom.mean())
dataset['Car']=dataset['Car'].fillna(dataset.Car.mean())
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset['YearBuilt']=dataset['YearBuilt'].fillna(0)

```

```python
dataset.isna().sum()
```

```python
##### drop all rows with missing values:
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset.info()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
### splitting data into feature set and dependent variable set:
X=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=2)
```

```python
####importing the linear regression models:
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
#### fit the training data to the model:
reg.fit(X_train,y_train)
```

```python
pred_reg=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,pred_reg)
r2
```

```python
###finding out the training set accuracy:
reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=60)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=50)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```
